# 01. Sigmoid Activation Function

#### Disply Equation 

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(6, 2))
plt.text(0.5, 0.5, r"$f(x) = \frac{1}{1 + e^{-x}}$", fontsize=20, ha='center', va='center')
plt.axis('off')  # Hide axes
plt.title('Sigmoid Activation Function Equation', fontsize=14)
plt.show()
```

#### Activation Function

```python
import numpy as np

def sigmoid(x): 
    return 1.0 / (1.0 + np.exp(-x)) 

inputs = [14, 4, -13] 

for x in inputs:
    print('{} -- > {}'.format(x, sigmoid(x))) 
```

#### Display Visually

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x): 
    return 1.0 / (1.0 + np.exp(-x)) 

inputs = [14, 4, -13]
outputs = [sigmoid(x) for x in inputs]

plt.figure(figsize=(8, 6))
plt.scatter(inputs, outputs, color='red', label='Input Points')
plt.plot(np.linspace(-15, 15, 300), sigmoid(np.linspace(-15, 15, 300)), color='blue', label='Sigmoid Curve')
plt.title('Sigmoid Function Visualization', fontsize=14)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Sigmoid Output', fontsize=12)
plt.axhline(0.5, color='gray', linestyle='--', label='y=0.5')
plt.axvline(0, color='gray', linestyle='--', label='x=0')
plt.grid(alpha=0.5)
plt.legend(fontsize=10)
plt.show()
```

# 02. Tanh Activation Function

#### Display Equation 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 2))
plt.text(0.5, 0.5, r"$f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$", fontsize=20, ha='center', va='center')
plt.axis('off')  # Hide axes
plt.title('Tanh Activation Function Equation', fontsize=14)
plt.show()
```

#### Activation Function

```python
import numpy as np 

def tanh(x):  
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)) 

inputs = [14, 4, -13]  

for x in inputs:  
    print('{} -- > {}'.format(x, tanh(x)))
```

#### Display Visually

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):  
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)) 

# Inputs
inputs = [14, 4, -13]
outputs = [tanh(x) for x in inputs]

x_vals = np.linspace(-15, 15, 300)
y_vals = tanh(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, color='blue', label='Tanh Curve')
plt.scatter(inputs, outputs, color='red', label='Input Points', zorder=5)
for i, val in enumerate(inputs):
    plt.text(val, outputs[i], f'({val}, {outputs[i]:.2f})', fontsize=10, ha='right')

plt.title('Tanh Function Visualization', fontsize=14)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Tanh Output', fontsize=12)
plt.axhline(0, color='gray', linestyle='--', label='y=0')
plt.axvline(0, color='gray', linestyle='--', label='x=0')
plt.grid(alpha=0.5)
plt.legend(fontsize=10)
plt.show()
```

# 03. ReLU Activation Function 

#### Display Equation 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 2))
plt.text(0.5, 0.5, r"$f(x) = \max(0, x)$", fontsize=20, ha='center', va='center')
plt.axis('off')  # Hide axes
plt.title('ReLU Activation Function Equation', fontsize=14)
plt.show()
```

#### Activation Function

```python
def relu(x):  
    return max(0.0, x) 

inputs = [14, 4, -13] 

for x in inputs: 
    print('{} -- > {}'.format(x, relu(x)))
```

#### Display Visually

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):  
    return max(0.0, x) 

inputs = [14, 4, -13]
outputs = [relu(x) for x in inputs]

x_vals = np.linspace(-15, 15, 300)
y_vals = [relu(x) for x in x_vals]

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, color='blue', label='ReLU Curve')
plt.scatter(inputs, outputs, color='red', label='Input Points', zorder=5)
for i, val in enumerate(inputs):
    plt.text(val, outputs[i], f'({val}, {outputs[i]:.2f})', fontsize=10, ha='right')

plt.title('ReLU Function Visualization', fontsize=14)
plt.xlabel('Input', fontsize=12)
plt.ylabel('ReLU Output', fontsize=12)
plt.axhline(0, color='gray', linestyle='--', label='y=0')
plt.axvline(0, color='gray', linestyle='--', label='x=0')
plt.grid(alpha=0.5)
plt.legend(fontsize=10)
plt.show()
```

# 04. Leaky ReLU Activation Function 

#### Display Equation 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 2))
plt.text(0.5, 0.5, r"$f(x) = \max(\alpha x, x)$", fontsize=20, ha='center', va='center')
plt.axis('off')  # Hide axes
plt.title('Leaky ReLU Activation Function Equation', fontsize=14)
plt.show()
```

#### Activation Function

```python
def leakyRelu(x):   
    return max(0.1*x, x) 

inputs = [14, 4, -13] 

for x in inputs:   
    print('{} -- > {}'.format(x, leakyRelu(x)))
```

#### Display Visually

```python
import numpy as np
import matplotlib.pyplot as plt

def leakyRelu(x):   
    return max(0.1 * x, x) 

inputs = [14, 4, -13]
outputs = [leakyRelu(x) for x in inputs]

x_vals = np.linspace(-15, 15, 300)
y_vals = [leakyRelu(x) for x in x_vals]

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, color='blue', label='Leaky ReLU Curve')
plt.scatter(inputs, outputs, color='red', label='Input Points', zorder=5)
for i, val in enumerate(inputs):
    plt.text(val, outputs[i], f'({val}, {outputs[i]:.2f})', fontsize=10, ha='right')

plt.title('Leaky ReLU Function Visualization', fontsize=14)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Leaky ReLU Output', fontsize=12)
plt.axhline(0, color='gray', linestyle='--', label='y=0')
plt.axvline(0, color='gray', linestyle='--', label='x=0')
plt.grid(alpha=0.5)
plt.legend(fontsize=10)
plt.show()
```

# 05. Softmax Activation Function 

#### Display Equation 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 2))
plt.text(0.5, 0.5, r"$f(x) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$", fontsize=20, ha='center', va='center')
plt.axis('off')  # Hide axes
plt.title('Softmax Activation Function Equation', fontsize=14)
plt.show()
```

#### Activation Function

```python
import numpy as np  

def softmax(x): 
    return np.exp(x) / np.exp(x).sum()  

inputs = [2, 4, -3] 

print(softmax(inputs))
```

#### Display Visually

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x): 
    exp_values = np.exp(x)
    return exp_values / exp_values.sum()

inputs = [2, 4, -3]
outputs = softmax(inputs)

x_vals = np.array(inputs)
y_vals = softmax(x_vals)

plt.figure(figsize=(8, 6))
plt.bar(x_vals, y_vals, color='blue', alpha=0.7, label='Softmax Output')
plt.scatter(x_vals, y_vals, color='red', label='Input Points', zorder=5)

for i, val in enumerate(x_vals):
    plt.text(val, y_vals[i], f'{y_vals[i]:.2f}', fontsize=10, ha='center')

plt.title('Softmax Function Visualization', fontsize=14)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Softmax Output', fontsize=12)
plt.grid(alpha=0.5)
plt.legend(fontsize=10)
plt.show()
```
